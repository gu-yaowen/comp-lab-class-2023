```python
import numpy as np
import mdtraj as md
from matplotlib import pyplot as plt
```

```python
meta_hill = np.loadtxt('../Inputs/dialaA/metad_hill_A_all.dat', comments='#')
plt.plot(meta_hill[:,0], meta_hill[:,1], alpha=0.5)
plt.xlabel('phi dihedral')
plt.ylabel('free energy (kcal/mol)')
plt.title('Estimate of the free energy as a function of the dihedral \n'
          'phi from a 10ns-long well-tempered metadynamics simulation.')
plt.savefig('../Figures/Lugano_ex2_phi-FE_A.png', dpi=300)
```

```python
# 0.1 ns
meta_hill_01 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat0.dat', comments='#')
# 0.3 ns
meta_hill_03 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat3.dat', comments='#')
# 1 ns
meta_hill_1 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat10.dat', comments='#')
# 2 ns
meta_hill_2 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat20.dat', comments='#')
# 5 ns
meta_hill_5 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat50.dat', comments='#')
# 10 ns
meta_hill_10 = np.loadtxt('../Inputs/dialaA/metad_hill_A.dat100.dat', comments='#')

plt.plot(meta_hill_01[:,0], meta_hill_01[:,1], alpha=0.5, label='0.1 ns')
plt.plot(meta_hill_03[:,0], meta_hill_03[:,1], alpha=0.5, label='0.3 ns')
plt.plot(meta_hill_1[:,0], meta_hill_1[:,1], alpha=0.5, label='1 ns')
plt.plot(meta_hill_2[:,0], meta_hill_2[:,1], alpha=0.5, label='2 ns')
plt.plot(meta_hill_5[:,0], meta_hill_5[:,1], alpha=0.5, label='5 ns')
plt.plot(meta_hill_10[:,0], meta_hill_10[:,1], alpha=0.5, label='10 ns')

plt.xlabel('phi dihedral')
plt.ylabel('free energy (kcal/mol)')
plt.legend()
plt.title('Estimates of the free energy as a function of the dihedral \n'
          'phi calculated every 100 Gaussian kernels deposited.')
plt.savefig('../Figures/Lugano_ex2_phi-FE_multiple_A.png', dpi=300)
```
